```python
# To clean the Place of Publication field, 
# we can combine Pandas str methods with NumPy’s np.where function
```

```python
# np.where(condition, then, else) --> this can be chained
```

```python
df['Place of Publication'].head(10)
```

```python
df.loc[4157862]
```

```python
df.loc[4159587]
```

```python
# the above contain same place of publication
# one with hypen another one does not contain hypen
```

```python
# To clean this column in one sweep, we can use str.contains() to get a boolean mask
```

```python
pub = df['Place of Publication']
```

```python
london = pub.str.contains('London')
london[:5]
```

```python
oxford = pub.str.contains('Oxford')
```

```python
df['Place of Publication'] = np.where(london, 'London',
                                      np.where(oxford, 'Oxford',
                                               pub.str.replace('-', ' ')))
df['Place of Publication'].head()
```
